```python
import os
from google.colab import userdata
from google import genai

# FIX: Import types directly from the genai module
from google.genai import types

try:
    api_key = userdata.get('GEMINI_API_KEY')
    client = genai.Client(api_key=api_key)
except Exception as e:
    print("⚠️ Key Error: Make sure your key is named GEMINI_API_KEY in the Secrets tab!")

SYSTEM_PROMPT = """
You are EcoMindAI, an expert AI Sustainability Assistant. Your goal is to provide
practical, accurate, and encouraging advice on eco-friendly living, sustainability,
waste reduction, and climate science.

Keep your answers scannable, actionable, and positive. If a user asks a question completely
unrelated to sustainability or science, gently guide the conversation back to environmental topics.
"""

def launch_colab_chatbot():
    print("🌿 EcoMindAI Engine Initialized in Colab 🌿")
    print("Type your question below. Type 'quit' or 'exit' to stop the session.\n")

    # FIX: Updated to use types.GenerateContentConfig
    config = types.GenerateContentConfig(
        system_instruction=SYSTEM_PROMPT,
        temperature=0.7
    )
    chat = client.chats.create(model="gemini-2.5-flash", config=config)

    while True:
        try:
            user_input = input("You: ")
            if user_input.lower() in ['quit', 'exit']:
                print("\nEcoMindAI: Small steps lead to big changes. Goodbye! 🌍")
                break
            if not user_input.strip():
                continue

            response = chat.send_message(user_input)
            print(f"EcoMindAI: {response.text}\n")
            print("-" * 40)

        except Exception as e:
            print(f"\nAn error occurred: {e}\n")
            break

# Run the chatbot
launch_colab_chatbot()
```

```python
import os
from google.colab import userdata
from google import genai
from google.genai import types

# ---------------------------------------------------------
# 1. INITIALIZE API CLIENT
# ---------------------------------------------------------
client = None
try:
    api_key = userdata.get('GEMINI_API_KEY')
    client = genai.Client(api_key=api_key)
except Exception as e:
    print("❌ API KEY ERROR: Check your 'GEMINI_API_KEY' in the Secrets tab!")

# ---------------------------------------------------------
# 2. FEATURE: CARBON FOOTPRINT CALCULATOR ENGINE
# ---------------------------------------------------------
def calculate_eco_score(transport, diet, waste):
    """Calculates an approximate annual kg CO2 score based on user habits"""
    score = 0

    # Transport calculation
    if transport == "Car (Gasoline)": score += 4600
    elif transport == "Electric Vehicle": score += 2000
    elif transport == "Public Transit": score += 800
    elif transport == "Walking/Biking": score += 0

    # Diet calculation
    if diet == "Heavy Meat Eater": score += 3000
    elif diet == "Balanced / Low Meat": score += 1700
    elif diet == "Vegetarian": score += 1400
    elif diet == "Vegan": score += 1000

    # Waste calculation
    if waste == "No recycling / Lots of waste": score += 1200
    elif waste == "Recycle sometimes": score += 800
    elif waste == "Strict recycling / Zero waste lifestyle": score += 300

    return score

# ---------------------------------------------------------
# 3. FEATURE: MULTIMODAL RECYCLING IDENTIFIER ENGINE
# ---------------------------------------------------------
def analyze_waste_image(image_url):
    """Downloads an image from a URL and asks Gemini if it's recyclable"""
    if not image_url or image_url == "Paste image URL here to test recycling scanner":
        return "No image URL provided to scanner."

    import requests
    from io import BytesIO

    try:
        print("🔍 Fetching and analyzing image...")
        response = requests.get(image_url, timeout=10)
        image_bytes = BytesIO(response.content).getvalue()

        # Structure the image data for the GenAI SDK
        image_part = types.Part.from_bytes(
            data=image_bytes,
            mime_type="image/jpeg"
        )

        prompt = """
        Analyze this image object. Tell me:
        1. What is this object?
        2. Is it recyclable, compostable, or trash?
        3. Give 1 clear instruction on how to properly dispose of or prepare it (e.g., rinse it, remove the cap).
        Keep your response brief and scannable.
        """

        # Use gemini-2.5-flash to look at the image data
        analysis = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=[image_part, prompt]
        )
        return analysis.text
    except Exception as e:
        return f"Could not analyze image. Error: {e}"

# ---------------------------------------------------------
# 4. COLAB FORMS INTERFACE (The User Interface)
# ---------------------------------------------------------
#@title 🌿 EcoMindAI Control Panel { display-mode: "form" }

#@markdown ### 💬 Ask the AI Chatbot
Chat_Question = "How can I reduce food waste at home?" #@param {type:"string"}

#@markdown ---
#@markdown ### 🚗 Carbon Footprint Calculator
Primary_Transportation = "Car (Gasoline)" #@param ["Car (Gasoline)", "Electric Vehicle", "Public Transit", "Walking/Biking"]
Diet_Type = "Balanced / Low Meat" #@param ["Heavy Meat Eater", "Balanced / Low Meat", "Vegetarian", "Vegan"]
Waste_Habits = "No recycling / Lots of waste" #@param ["No recycling / Lots of waste", "Recycle sometimes", "Strict recycling / Zero waste lifestyle"]

#@markdown ---
#@markdown ### 📸 Image Recycling Scanner
Waste_Image_URL = "Paste image URL here to test recycling scanner" #@param {type:"string"}

def run_ecomind_app():
    if client is None:
        print("🛑 Cannot run. Please check your API key setup.")
        return

    print("=" * 60)
    print("🌿 EcoMindAI MULTI-FEATURE DASHBOARD RESULTS 🌿")
    print("=" * 60)

    # Run Carbon Calculator
    eco_impact = calculate_eco_score(Primary_Transportation, Diet_Type, Waste_Habits)
    print(f"\n📊 [CARBON CALCULATOR RESULTS]")
    print(f" -> Your estimated carbon footprint: {eco_impact} kg CO2 / year.")
    if eco_impact > 6000:
        print("💡 Tip: Switching to public transit or eating more plant-based meals can drastically drop your score!")
    elif eco_impact > 3000:
        print("💡 Tip: You're doing great! Try focusing on reducing packaging waste to lower it further.")
    else:
        print("💡 Tip: Exceptional eco-friendly footprint! Keep leading by example.")
    print("-" * 40)

    # Run Image Scan if a URL was provided
    if Waste_Image_URL and Waste_Image_URL != "Paste image URL here to test recycling scanner":
        print(f"\n📸 [IMAGE RECYCLING SCANNER RESULTS]")
        image_result = analyze_waste_image(Waste_Image_URL)
        print(image_result)
        print("-" * 40)

    # Run Chatbot Response
    if Chat_Question:
        print(f"\n💬 [CHATBOT RESPONSE]")
        print(f"You asked: '{Chat_Question}'\n")

        system_instruction = "You are EcoMindAI, an expert sustainability assistant. Give practical, scannable eco advice."
        response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=Chat_Question,
            config=types.GenerateContentConfig(system_instruction=system_instruction)
        )
        print(f"EcoMindAI: {response.text}")
        print("=" * 60)

# Execute the application functions
run_ecomind_app()
```

```python
import os
import streamlit as st
from google import genai
from google.genai import types

# 1. SETUP & SAFETY CHECK
try:
    # Looks for GEMINI_API_KEY environment variable automatically
    client = genai.Client()
except Exception as e:
    st.error("❌ API Connection Failed! Make sure GEMINI_API_KEY is set in your system environment.")
    st.stop()

# 2. CONFIGURE WEB PAGE
st.set_page_config(page_title="EcoMindAI - Assistant", page_icon="🌿", layout="centered")

st.title("🌿 EcoMindAI: Sustainability Assistant")
st.caption("🚀 Final Year Project Prototype | Powered by Gemini 2.5 Flash")
st.write("Ask anything about eco-friendly living, zero-waste strategies, or recycling guidelines.")

# 3. SET SYSTEM PERSONA
SYSTEM_PROMPT = """
You are EcoMindAI, an expert AI Sustainability Assistant. Your goal is to provide
practical, accurate, and encouraging advice on eco-friendly living, sustainability,
waste reduction, and climate science.

Keep your answers scannable, using short bullet points where appropriate. Always maintain a positive,
helpful peer tone. If a user asks a question completely unrelated to sustainability, environment, or science,
gently bring the conversation back to environmental topics.
"""

# 4. INITIALIZE CHAT MEMORY (Streamlit Session State)
if "messages" not in st.session_state:
    # Start with a clean history and a welcoming bot message
    st.session_state.messages = [
        {"role": "assistant", "content": "Hello! I am EcoMindAI. Ask me anything about sustainability or waste management!"}
    ]

# 5. RENDER THE HISTORICAL CONVERSATION
for msg in st.session_state.messages:
    with st.chat_message(msg["role"]):
        st.markdown(msg["content"])

# 6. CAPTURE NEW USER INPUT
if user_prompt := st.chat_input("Type your environmental question here..."):

    # Render user message instantly on screen
    with st.chat_message("user"):
        st.markdown(user_prompt)

    # Store user message in history memory
    st.session_state.messages.append({"role": "user", "content": user_prompt})

    # 7. GENERATE RESPONSES WITH CONTEXT HISTORY
    with st.chat_message("assistant"):
        with st.spinner("EcoMindAI is thinking..."):
            try:
                # Convert our storage format into Google API content history format
                formatted_history = []
                for msg in st.session_state.messages[:-1]: # Exclude the very last prompt we just added
                    formatted_history.append(
                        types.Content(
                            role=msg["role"],
                            parts=[types.Part.from_text(text=msg["content"])]
                        )
                    )

                # Start or update the multi-turn conversation session
                chat_session = client.chats.create(
                    model="gemini-2.5-flash",
                    history=formatted_history,
                    config=types.GenerateContentConfig(
                        system_instruction=SYSTEM_PROMPT,
                        temperature=0.7
                    )
                )

                # Send the newest prompt through the active chat thread
                response = chat_session.send_message(user_prompt)

                # Display output and save to persistent session memory
                st.markdown(response.text)
                st.session_state.messages.append({"role": "assistant", "content": response.text})

            except Exception as e:
                st.error(f"Error generating AI response: {e}")
```

```python
import os
import streamlit as st
from google import genai
from google.genai import types

# 1. SETUP & SAFETY CHECK
try:
    client = genai.Client()
except Exception as e:
    st.error("❌ API Connection Failed! Make sure GEMINI_API_KEY is set in your system environment.")
    st.stop()

# 2. CONFIGURE WEB PAGE
st.set_page_config(page_title="EcoMindAI - Assistant", page_icon="🌿", layout="centered")

st.title("🌿 EcoMindAI: Sustainability Assistant")
st.caption("🚀 Final Year Project Prototype | Powered by Gemini 2.5 Flash")

# 3. INITIALIZE LIVE CHAT SESSION OBJECT (The Speed Secret)
SYSTEM_PROMPT = "You are EcoMindAI, an expert AI Sustainability Assistant. Give fast, practical, scannable eco advice."

if "chat_session" not in st.session_state:
    # We build the persistent connection object ONCE. It manages its own history instantly.
    st.session_state.chat_session = client.chats.create(
        model="gemini-2.5-flash",
        config=types.GenerateContentConfig(
            system_instruction=SYSTEM_PROMPT,
            temperature=0.7
        )
    )
    # Visual message storage tracker for UI rendering
    st.session_state.ui_messages = [
        {"role": "assistant", "content": "Hello! I am EcoMindAI. Ask me anything about sustainability or waste management!"}
    ]

# 4. RENDER PRESENTATION CHAT UI
for msg in st.session_state.ui_messages:
    with st.chat_message(msg["role"]):
        st.markdown(msg["content"])

# 5. CAPTURE & PROCESS USER INPUT
if user_prompt := st.chat_input("Type your environmental question here..."):

    # Render user prompt instantly
    with st.chat_message("user"):
        st.markdown(user_prompt)
    st.session_state.ui_messages.append({"role": "user", "content": user_prompt})

    # Fetch response via the active stream link
    with st.chat_message("assistant"):
        with st.spinner("Thinking..."):
            try:
                # Direct stream execution - no history rebuilding required!
                response = st.session_state.chat_session.send_message(user_prompt)
                st.markdown(response.text)
                st.session_state.ui_messages.append({"role": "assistant", "content": response.text})
            except Exception as e:
                st.error(f"Error: {e}")
```
